```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
%matplotlib inline
%matplotlib notebook
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sinasc = pd.read_csv('SINASC_RO_2019.csv')
sinasc.head()
```

```python
idade = sinasc['IDADEMAE']
plt.boxplot(idade)
plt.title('IDADE DA MÃE')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.close('all')

fig = plt.figure()
plt.title('PESO DO BEBÊ')
peso = sinasc['PESO']
plt.boxplot(peso)
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.close('all')

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.hist(sinasc['IDADEMAE'],rwidth=0.7)
ax1.set_title('IDADE DA MÃE')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.close('all')

fig = plt.figure()
ax2 = fig.add_subplot(1,1,1)
ax2.hist(sinasc['PESO'],rwidth=0.7)
ax2.set_title('PESO DO BEBÊ')
plt.show()
```
